```python
# suprimer les colonnes manquantes des variables qualitatifs et pour les variables qualitatives remplacer les données manquantes par la valeur de la ligne précédente

df.dropna(subset= ['Sexe', 'Situation_familiale','Travailleur_independant'], inplace = True)
```

```python
df
```

```python
# remplacer chaque valeur manquantes par la valeur precédente

df.fillna(method = 'bfill', inplace = True)
```

```python
df.isnull().sum()
```

```python
# remplacer MAL par 1 et FEMELLE par 0

df['Sexe'].replace(['Male', 'Female'], [1, 0], inplace = True)
```

```python
df
```

```python
# Remplacer yes et no par 0 et 1 sur le col Situation_familiale
# De mm pour Travailleur_independant
df['Situation_familiale'].replace(['Yes', 'No'], [1, 0], inplace = True)
df['Travailleur_independant'].replace(['Yes', 'No'], [1, 0], inplace = True)
```

```python
df
```

```python
# Remplacer graduate et no no graduate 0 et 1 sur le col Diplome
df['Diplome'].replace(['Graduate', 'Not Graduate'], [1, 0], inplace = True)
```

```python
df
```

```python
# urbain
df['Residence'].replace(['Urban', 'Semiurban', 'Rural'], [2, 1, 0], inplace = True)
```

```python
df
```

```python
# Décision
df['Decision'].replace(['Y', 'N'], [1, 0], inplace = True)
```

```python
df
```

```python
#  suprimer le colonne client vu qu'il n'a pas d'impact
df.drop('Client_ID', axis = 1, inplace = True)
```

```python
df
```

```python
# afficher les columns
df.columns
```

```python
# séparation des données : variables explcatives et variables à expliquées
```

```python
X = df[['Sexe', 'Situation_familiale', 'Personnes_charge', 'Diplome',
       'Travailleur_independant', 'Revenu_imposable', 'Revenu_forfitaire',
       'Montant_pret', 'Duree_pret', 'Hitorique_credit', 'Residence',]]
X
```

```python
y = df[['Decision']]
y
```

```python
# décomposition des données: données d'entrainement et données de test

from sklearn.model_selection import train_test_split
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 101)
print("données entrainement", X_train.shape)
print("données de test", X_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# creer un model
model = KNeighborsClassifier()

# entrainer le model a travers fit
model.fit(X_train, y_train)
```

```python
# prédiction

predictions = model.predict(X_train)
predictions
```

```python
# calculer le model à partir de X_train et y_train

print("performance du model sur la base des données apprentissage", model.score(X_train, y_train))  #la performance du model est de 0.74
```

```python
# de mm pour le X_test et y_test

print("performance du model sur la base des données apprentissage", model.score(X_test, y_test))

```

```python
# optimiser le model par GridSearchCV

from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'n_neighbors': np.arange(1,26),
              'weights': ['uniform', 'distance'],
              'metric': ['euclidean', 'manhattan', 'minkovski']}
```

```python
grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv = 10)
```

```python
# entrainer le model

grid.fit(X_train, y_train)
```

```python
# le meilleur score

grid.best_score_           # per formance du meilleur model
```

```python
#meilleur parametre

grid.best_params_

```

```python
# sauvegarder le model

model = grid.best_estimator_
model
```

```python
# afficher le score

model.score(X_test, y_test)
```

```python
# creer un datafram à partir de ces données

df = pd.DataFrame(cancer['data'], columns= cancer['feature_names'])
```

```python
df
```

```python
# importer un modul

from sklearn.preprocessing import StandardScaler
```

```python
data_scaler = StandardScaler()
data_scaler.fit(df)
```

```python
data_scaler
```

```python
data_scaler = data_scaler.transform(df)
```

```python
data_scaler                    # on n'a obtenu une matrice NUMPY
```

```python

# creer une dataframe
df2 = pd.DataFrame(data_scaler, columns= cancer['feature_names'])
```

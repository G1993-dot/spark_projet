```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ARBRE DE DECISION
```

```python
df = pd.read_csv('/content/dataCancer.csv')
df
```

```python
df.isnull().sum()
```

```python
df.diagnosis.value_counts()
```

```python
df.columns
```

```python
df.drop(['id', 'Unnamed: 32'], axis = 1, inplace = True)
```

```python
df['diagnosis'].replace(['M', 'B'], [0, 1], inplace = True)
```

```python
df
```

```python
X = df[['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]
```

```python
X
```

```python
y = df['diagnosis']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state= 100)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.tree import DecisionTreeClassifier        
```

```python
model = DecisionTreeClassifier()          # model arbre de décision
```

```python
model.fit(X_train, y_train)
```

```python
predict = model.predict(X_train)
predict
```

```python
model.score(X_train, y_train)       # le score est de 100%
```

```python
model.score(X_test, y_test)       # score sur lles test
```

```python
# gerer un raport de classification
from sklearn.metrics import classification_report
```

```python
print(classification_report(X_test, predict))   # classification de performance du model
```

```python
# visualiser l'arbre décisionnel
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize = (80, 65))
plot_tree(model, feature_names=X.columns, class_names= ['M', 'B'], filled=True)
```
